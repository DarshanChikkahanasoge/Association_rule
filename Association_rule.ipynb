```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df=pd.read_excel('../data/data_association_rules.xlsx')
```

```python
df.shape
```

```python
df.head()
```

```python
df['quantity']=1
```

```python
df.head()
```

```python
df.groupby(['Transaction']).count()
```

```python
matrix = df.pivot_table(index='Transaction',columns='Items',values='quantity')
```

```python
matrix
```

```python
matrix1=matrix.applymap(lambda x:1 if x>0 else 0)
```

```python
matrix1
```

```python
itemsets=apriori(matrix1,
                min_support=0.4,
                use_colnames=True)
```

```python
itemsets
```

```python
association=association_rules(itemsets,
                             metric="confidence",
                             min_threshold=0.8)
```
